## Static Web Scraping 

Resource:

- [Project1_bs4_only](https://zhuanlan.zhihu.com/p/355013457) 

For static web sraping. Its very simple with only requests and bs4 needed 



```python
age=2
url="https://www.rong360.com/gl/xinshoudaikuangonglue/list_20_{}.html".format(age)
home='https://www.rong360.com'
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/51.0.2704.63 Safari/537.36'}
req=requests.get(url,headers=headers)
soup=BeautifulSoup(req.content,'lxml')
H3=soup.find_all('h3')
names=[]
relates=[]
for hi in H3:
    name=hi.text.strip()
    names.append(name)
    urli=hi.a.get('href')
    url=home+urli
    reqi=requests.get(url,headers=headers)
    soupi=BeautifulSoup(reqi.content,'lxml')
    contenti=soupi.find('div',class_='act-content')
    ul=contenti.find('ul',class_='related-articles')
    lis=ul.find_all('li')
    relate=[]
    for li in lis[1:]:
        relate.append(li.text.strip())
    relates.append(relate)
```

```python
pd.DataFrame({'Article':names,'Related':relates})
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
driver.get('https://forums.edmunds.com/discussion/2864/general/x/entry-level-luxury-performance-sedans/p2')
```

```python
# Find XPATH:
//*[@id="Comment_1738047"]/div/div[2]/div[1]/span[1]/a[2]
//*[@id="Comment_1738152"]/div/div[2]/div[1]/span[1]/a[2]
```

```python
userid_element =driver.find_element(By.XPATH, '//*[@id="Comment_1738047"]/div/div[2]/div[1]/span[1]/a[2]')
userid=userid_element.text
```

```python
date_element=driver.find_element(By.XPATH,'//*[@id="Comment_1738047"]/div/div[2]/div[2]/span/a/time')
date_element.get_attribute('title')
```

```python
comment_element=driver.find_element(By.XPATH,'//*[@id="Comment_1738047"]/div/div[3]/div/div[1]')
comment=comment_element.text
```

```python
# RE
userids_elements = driver.find_elements(By.XPATH,'//*[contains(@id,"Comment_")]')
comment_ids=[]
for i in userids_elements:
    comment_ids.append(i.get_attribute('id'))
```

```python
names=[]
dates=[]
comments=[]
for i in  comment_ids:
    userid_element =driver.find_element(By.XPATH, '//*[@id="'+i+'"]/div/div[2]/div[1]/span[1]/a[2]')
    userid=userid_element.text
    names.append(userid)
    
    date_element=driver.find_element(By.XPATH,'//*[@id="'+i+'"]/div/div[2]/div[2]/span/a/time')
    date=date_element.get_attribute('title')
    dates.append(date)
    
    comment_element=driver.find_element(By.XPATH,'//*[@id="'+i+'"]/div/div[3]/div/div[1]')
    comment=comment_element.text
    comments.append(comment)
```

```python
import pandas as pd
data=pd.DataFrame({'Name':names,'Dates':dates,'Comments':comments})
data.head()
```
