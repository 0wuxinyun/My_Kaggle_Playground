## Basic Data Exploration

```python
import pandas as pd
```

```python
path = "./melb_data.csv"
data=pd.read_csv(path)
data.head()
```

```python
# statistical summary of the data :
data.describe()
```

```python
# missing values 
print(data.columns)
data.dropna(axis=0).head()
```

```python
# selection:
price=data.Price
price[:4]
features = ['Rooms', 'Bathroom', 'Landsize', 'Lattitude', 'Longtitude']
d_f=data[features]
d_f.describe()
```

## Modling
- Define: What type of model will it be? A decision tree? Some other type of model? Some other parameters of the model type are specified too.
- Fit: Capture patterns from provided data. This is the heart of modeling.
- Predict: Just what it sounds like
- Evaluate: Determine how accurate the model's predictions are.

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split as split
# Define model. Specify a number for random_state to ensure same results each run
train_X, val_X, train_y, val_y = split(d_f,price,random_state=0)
model = DecisionTreeRegressor(random_state=1)
model.fit(train_X,train_y)
```

```python
## model evaluation
from sklearn.metrics import mean_absolute_error as mae
predictions=model.predict(val_X)
mae(val_y,predictions)
```

## Better model to overcome overfittig & underfitting

```python
from sklearn.ensemble import RandomForestRegressor as rfr
model2=rfr(random_state=1)
model2.fit(train_X,train_y)
```

```python
predictions2=model2.predict(val_X)
mae(val_y,predictions2)
```
